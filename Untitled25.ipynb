```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

a=pd.read_csv('/content/placement.csv')

plt.scatter(a['cgpa'],a['package'])
plt.xlabel('cgpa')
plt.ylabel('package')
plt.title('scatter plot')
plt.show()
```

```python
a
x=a.iloc[:,0:1]
y=a.iloc[:,-1]
print(type(x))
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
model.predict([[6.89]])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

a=pd.read_csv('/content/placement.csv')

plt.scatter(a['cgpa'],a['package'])
plt.plot(x,model.predict(x),color='red')
plt.xlabel('cgpa')
plt.ylabel('package')
plt.title('scatter plot')
plt.show()
```
